# Implementación de una técnica de aprendizaje máquina sin el uso de un framework

```python
# Lectura de librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Lectura del dataset "Canada_per_capita_income (1).csv"
data = pd.read_csv("Canada_per_capita_income (1).csv")
datos = pd.DataFrame(data)
datos.head()
```

```python
# Impresion del dataset
datos
```

```python
#Se le asigna a la variable x la columna de 'year'
x = datos['year']
# Se le asigna a la variablr y la columna de 'income'
y = datos['income']

# Crear lista con los hiper-parámetros iniciales (thetas)
theta = [1, 1]
# Cargar el valor del learning rate (alpha) de 0.000001
alpha = 0.0000001
print(x.head())
print(y.head())
```

## Inicio de implementacion del algoritmo de Regresion Lineal

```python
# Crear función lambda para la función de hipótesis
h = lambda x, theta: theta[0] + theta[1] * x
# Calcular el total de muestras a partir de los datos (n)
n = len(x)
```

```python
for i in range(100):
  delta = []
  deltax = []
  for xi, yi in zip(x,y):
    pred = h(xi,theta)
    delta.append(pred-yi)
    deltax.append((pred-yi)*xi)

  dJt0 = 1/n*sum(delta)
  dJt1 = 1/n*sum(deltax)

  print(dJt0)
  print(dJt1)

    # Actualizar theta0
  theta[0] = theta[0] - alpha*dJt0

    # Actualizar theta1
  theta[1] = theta[1] - alpha*dJt1
  print(theta)
```

## Resultados

```python
#Se creara una lista con diferentes años para poder obtener los diferentes resultados y posteriormente de esos años los que se puedan buscar se hara para comprobar esos resultados
lista_valores = [1963, 1996, 2000, 2001, 2010, 2032, 2049]

#Se crea un for loop para poder pasar la lista con los diferentes valores a probar y conocer las predicciones de dichos años
for j in lista_valores:
    h = theta[0] + (theta[1]*j)
    print("Este es el año de interes:", j, "y este es su valor:", h)

# Verificacion de las predicciones ya existentes
lista_valores_existentes = [1963, 1996, 2000, 2001, 2010]
for k in lista_valores_existentes:
    print(datos[datos['year'].isin([k])])
```
