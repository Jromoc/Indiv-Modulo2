```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
datos = pd.read_csv("Canada_per_capita_income (1).csv")
```

```python
x = []
x = datos['year']

y = []

y = datos['income']

# Crear lista con los hiper-parámetros iniciales (thetas)
theta = [1, 1]
# Cargar el valor del learning rate (alpha)
alpha = 0.0000001
print(x.head())
print(y.head())
```

```python
# Crear función lambda para la función de hipótesis
h = lambda x, theta: theta[0] + theta[1] * x
# Calcular el total de muestras a partir de los datos (n)
n = len(x)
```

```python
for i in range(1000):
  delta = []
  deltax = []
  for xi, yi in zip(x,y):
    pred = h(xi,theta)
    delta.append(pred-yi)
    deltax.append((pred-yi)*xi)

  dJt0 = 1/n*sum(delta)
  dJt1 = 1/n*sum(deltax)

  print(dJt0)
  print(dJt1)

      # Actualizar theta0
  theta[0] = theta[0] - alpha*dJt0

    # Actualizar theta1
  theta[1] = theta[1] - alpha*dJt1
  print(theta)
```
